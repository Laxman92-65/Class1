```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# Setup headless Chrome
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Product URL
url = "https://www.flipkart.com/sony-bravia-2-138-8-cm-55-inch-ultra-hd-4k-led-smart-google-tv-feature-playstation-5/p/itmbcfe0b4a5dbbd?pid=TVSHYP2EHXJPCCSS"
driver.get(url)

# Let the page load
time.sleep(3)

# Close login popup if it appears
try:
    close_btn = driver.find_element(By.XPATH, "//button[contains(text(),'‚úï')]")
    close_btn.click()
except:
    pass

# Extract title
try:
    title = driver.find_element(By.CLASS_NAME, "B_NuCI").text.strip()
except:
    title = "Title not found"

# Extract price
try:
    price = driver.find_element(By.CLASS_NAME, "_30jeq3").text.strip()
except:
    price = "Price not found"

# Extract highlights
highlights = []
try:
    lis = driver.find_elements(By.CSS_SELECTOR, "div._2418kt li")
    highlights = [li.text.strip() for li in lis]
except:
    pass

# Print results
print("üõí Product Title:", title)
print("üí∞ Price:", price)
print("üîç Highlights:")
for h in highlights:
    print("   ‚Ä¢", h)

driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

# Target Snapdeal product URL
url = "https://www.snapdeal.com/product/zebronics-zebwarrior-20-speakers-black/644024582446#bcrumbLabelId:153"

# Spoof a real browser
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/114.0.0.0 Safari/537.36"
    )
}

# Request the page
response = requests.get(url, headers=headers)

# Basic check
print("Status Code:", response.status_code)  # Should be 200

# Parse HTML
soup = BeautifulSoup(response.content, "html.parser")

# Extract title
title_tag = soup.find("h1", class_="pdp-e-i-head")
title = title_tag.get_text(strip=True) if title_tag else "Title not found"

# Extract price
price_tag = soup.find("span", class_="payBlkBig")
price = price_tag.get_text(strip=True) if price_tag else "Price not found"

# Extract discount (if present)
disc_tag = soup.find("span", class_="pdp-discount")
discount = disc_tag.get_text(strip=True) if disc_tag else "No discount info"

# Availability or stock status
avail_tag = soup.find("div", class_="pdp-variation")
availability = avail_tag.get_text(strip=True) if avail_tag else "Availability unknown"

# Output results
print("üõí Product Title:", title)
print("üí∞ Price:", price)
print("üè∑Ô∏è Discount:", discount)
print("üì¶ Availability:", availability)

```
